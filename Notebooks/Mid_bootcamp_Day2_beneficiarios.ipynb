From the database generated yesterday, I generated a function to determinate the beneficiary population assigned from 0 to 4 years by UMF and answer the question if this data has changed along the last 10 years.
```python
nulls_df = pd.DataFrame(round(beneficiarios.isna().sum()/len(beneficiarios),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
display(nulls_df)
```

```python
d = beneficiarios.copy()
d = d[(d['st_tit_fam'] == 2) & (d['id_calidad'] >= 13) & (d['cve_rango_edad'].isin(['E0','E1','E2','E3'])) & (d['st_consultorio'] == 1)]
d_grouped = d.groupby(['period','name_entidad','name_delegacion','id_umf_rp', 'nombre_umf_rp','cve_sexo']).agg({'tot_casos':sum}).rename(columns={'tot_casos':'Total_casos'}).reset_index()
d_grouped
```
