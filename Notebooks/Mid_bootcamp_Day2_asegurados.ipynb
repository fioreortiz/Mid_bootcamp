These tables includes the description of cve_entidad and cve_municipio and are exported to different files that will be process for its presentation in tableau.

```python
a_sector = asegurados.groupby(['period','name_delegacion','name_entidad','sector_economico_1']).agg({'asegurados':sum,'ta':sum, 'tpu':sum}).rename(columns={'asegurados':'Total_asegurados','ta':'Total_ta','tpu':'Total_tpu'}).reset_index()
```
