```python
import pandas as pd
import seaborn as sns
```

```python
asegurados2013 = pd.read_csv('../Data/raw/asg_2013_01_31.csv', encoding="ISO-8859-1", sep="|", low_memory=False)
entidad = pd.read_excel('../Data/documentation/location_entidad.xlsx')
```

```python
def get_dict_from_entidad():
    d = {}
    for index, row in entidad.iterrows():
        id = row["cve_entidad"]
        if id not in d:
            d[id] = row["name_entidad"]
    return d
```

```python
d = get_dict_from_entidad()
print(d)
```

```python
def get_entidad_name(x):
    return d[x]
asegurados2013["name_entidad"] = asegurados2013["cve_entidad"].apply(get_entidad_name)
```
